## Draw Circle

Syntax: cv2.circle(image, center_coordinates, radius, color, thickness)
Parameters: 
image: It is the image on which circle is to be drawn. 

center_coordinates: It is the center coordinates of circle. The coordinates are represented as tuples of two values i.e. (X coordinate 
value, Y coordinate value). 

radius: It is the radius of circle. 

color: It is the color of border line of circle to be drawn. For BGR, we pass a tuple. eg: (255, 0, 0) for blue color. 

thickness: It is the thickness of the circle border line in px. Thickness of -1 px will fill the circle shape by the specified color.

Return Value: It returns an image. 

```python
x_center=int(image2.shape[0]/2)
y_center=int(image2.shape[1]/2)


img_c=cv2.circle(image2.copy(),(y_center,x_center),60,(255,10,50),5)
cv2.imshow("Circle",img_c)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Draw Text 

Syntax: cv2.putText(image, text, org, font, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])

Parameters:
image: It is the image on which text is to be drawn.

text: Text string to be drawn.

org: It is the coordinates of the bottom-left corner of the text string in the image. The coordinates are represented as tuples of two values i.e. (X coordinate value, Y coordinate value).

font: It denotes the font type. Some of font types are FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN, , etc.

fontScale: Font scale factor that is multiplied by the font-specific base size.

color: It is the color of text string to be drawn. For BGR, we pass a tuple. eg: (255, 0, 0) for bluecolor.

thickness: It is the thickness of the line in px.

lineType: This is an optional parameter.It gives the type of the line to be used.

bottomLeftOrigin: This is an optional parameter. When it is true, the image data origin is at the bottom-left corner. Otherwise, it is at the top-left corner.

Return Value: It returns an image.

```python
cv2.putText(image2,"AI",(100,x_center),cv2.FONT_HERSHEY_SIMPLEX,2,(255,100,50),2,cv2.LINE_AA)
cv2.imshow("Text",image2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2 
image=cv2.imread('first_im.jpg',1)

smaller1=cv2.resize(image,(150,150))

cv2.imshow("Ener Area image",smaller1 )
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt 

plt.imshow(cv2.cvtColor(smaller1,cv2.COLOR_BGR2RGB))
plt.title("INTER_AREA")
plt.show()
```

```python

"""
interpolation


cv2.INTER_AREA : this is used when you ned to shrink an image 
cv2.INTER_CUBIC : slow nut more efficient
cv2.INTER_LINEAR : this is used when zooming is required

"""

smaller1=cv2.resize(image,(150,150),interpolation=cv2.INTER_AREA)
smaller2=cv2.resize(image,(150,150),interpolation=cv2.INTER_CUBIC)
smaller3=cv2.resize(image,(150,150),interpolation=cv2.INTER_LINEAR)

title=['INTER_AREA','INTER_CUBIC','INTER_LINEAR']
Images=[smaller1,smaller2,smaller3]


for i in range(3):
    plt.subplot(2,2,i+1)
    plt.title(title[i])
    plt.imshow(cv2.cvtColor(Images[i],cv2.COLOR_BGR2RGB))
    
plt.show()
```

```python
image=cv2.imread('first_im.jpg',1)

b,g,r=cv2.split(image)
```

```python
plt.imshow(cv2.cvtColor(r,cv2.COLOR_GRAY2RGB))
plt.title("Red Channel")
plt.show()

plt.imshow(cv2.cvtColor(b,cv2.COLOR_GRAY2RGB))
plt.title("blue Channel")
plt.show()


plt.imshow(cv2.cvtColor(g,cv2.COLOR_GRAY2RGB))
plt.title("Green Channel")
plt.show()
```
